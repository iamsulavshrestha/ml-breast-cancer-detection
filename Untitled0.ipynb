## Splitting the dataset train and test set

```python
dataset.head()
```

```python
# matrix of feature / independent variables
x = dataset.iloc[:, 1:-1].values
```

```python
x.shape
```

```python
# target variable / dependent variable
y=dataset.iloc[:, -1].values
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Part 2: Building model

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifir_lr = LogisticRegression(random_state=0)
```

```python
classifir_lr.fit(x_train, y_train)
```

```python
y_pred = classifir_lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, precision_score, recall_score
```

```python
acc = accuracy_score(y_test, y_pred)
```

```python
f1 = f1_score(y_test, y_pred)
```

```python
prec = precision_score(y_test, y_pred)
```

```python
rec = recall_score(y_test, y_pred)
```

```python
results = pd.DataFrame([['Logistic Regression', acc, f1, prec, rec]],
                       columns = ['Model', 'Accuracy', 'F1 Score', 'Precision', 'Recall'])
```

```python
results
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

### Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracies = cross_val_score(estimator=classifir_lr, X=x_train, y=y_train, cv=10)
print("Accuracy is {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation is {:.2f} %".format(accuracies.std()*100))
```

# Part 3 : Randomized Search to find the best parameteres (Logistic Regression)

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
parameters = {'penalty':['l1','l2', 'elasticnet', 'none'],
              'C':[0.25, 0.5, 0.75, 1.0, 1.25, 1.5, 1.75, 2.0],
              'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
              }
parameters
```

```python
random_search = RandomizedSearchCV(
    estimator=classifir_lr, param_distributions=parameters, n_iter=10, scoring='roc_auc', n_jobs=-1, cv=10, verbose=3
)
```

```python
random_search.fit(x_train, y_train)
```

```python
random_search.best_estimator_
```

```python
random_search.best_score_
```

```python
random_search.best_params_
```

# Part 4: Final Model (Logistic Regression)

```python
from sklearn.linear_model import LogisticRegression
classifir = LogisticRegression(C=1.5, random_state=0, solver='saga') # pass params from result of line 42
classifir.fit(x_train, y_train)

y_pred = classifir.predict(x_test)
```

acc = accuracy_score(y_test, y_pred)

results

```python
from sklearn.model_selection import cross_val_score

accuracies = cross_val_score(estimator=classifir, X=x_train, y=y_train, cv=10)
print("Accuracy is {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation is {:.2f} %".format(accuracies.std()*100))
```
